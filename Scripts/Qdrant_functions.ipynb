```python
from qdrant_client import QdrantClient
from qdrant_client.models import Filter, FieldCondition, MatchValue

# Connect to Qdrant
client = QdrantClient(
    url="http://localhost:6333",  # change if using cloud
    # api_key="YOUR_API_KEY"      # uncomment for Qdrant Cloud
)

COLLECTION_NAME = "user_uploads"

# Define the payload filter
payload_filter = Filter(
    must=[
        FieldCondition(
            key="document_name",           # payload key
            match=MatchValue(value="Ankit Sharma")
        )
    ]
)

# Delete points matching the filter
client.delete(
    collection_name=COLLECTION_NAME,
    points_selector=payload_filter
)

print("✅ Deleted all chunks where document_name == 'Ankit Sharma'")
```
